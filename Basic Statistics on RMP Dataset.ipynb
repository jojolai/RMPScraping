```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

femaleProfOQRatings = []
maleProfOQRatings = []
with open('profData.csv',encoding = "ISO-8859-1") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
#         print(row[1])
        if line_count != 0:
            if(int(row[29]) < 10):
                break;
            if row[34] == "male":
                maleProfOQRatings.append(float(row[6]))
            else:
                femaleProfOQRatings.append(float(row[6]))
            line_count += 1
        else:
            line_count += 1
    print(f'Processed {line_count} lines.')
    

plt.hist(maleProfOQRatings, bins=8, range = (1,5))
plt.ylabel('Number of Ratings')
plt.xlabel('Overall Quality Rating')
plt.title('Overall Quality Rating for Male Professors at BU')
plt.show()

plt.hist(femaleProfOQRatings, bins=8, range = (1,5))
plt.ylabel('Number of Ratings')
plt.xlabel('Overall Quality Rating')
plt.title("Overall Quality Rating for Female Professors at BU")
plt.show()

```
